```python
import numpy as np
import scipy.fftpack as spfft
import matplotlib.pyplot as plt
```

```python
# Defining a function to Fourier transform

Fs = 150.0; # sampling rate
Ts = 1.0/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector

ff = 5; # frequency of the signal
y = np.sin(2*np.pi*ff*t) # example signal to be Fourier transformed

frq = spfft.fftfreq(t.shape[-1])*Fs # frequency spectrum, in standard order (zero, positive, then negative frequencies)

Y = spfft.fft(y) # fft computing
```

```python
# Plotting the original function and its Fourier transform

fig = plt.figure() # instantiate figure object
ax1 = fig.add_subplot(2, 1, 1) # subplot environment has 2 rows, 1 column, ax1 in 1st position
ax2 = fig.add_subplot(2, 1, 2) # subplot environment has 2 rows, 1 column, ax2 in 2nd position

ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude')
ax1.plot(t,y)
ax2.set_xlabel('Frequency (Hz)')
ax2.set_ylabel('|Y(Frequency)|')
ax2.plot(frq,abs(Y),'r') # plotting the spectrum
#ax2.set_xlim(-20,20)
fig.subplots_adjust(hspace=0.5) # height adjustment so ax1 xlabel is visible

plt.show()
```
